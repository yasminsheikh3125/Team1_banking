```python
#ploting the boxplot of age column with respect to churn

plt.figure(figsize=(5,5))

sns.boxplot(x='churn', y='age', data=df)
# ploting the tiltle and lables of the boxplot

plt.title("Churn vs Age")
plt.xlabel("Churn (0 = No, 1 = Yes)")
plt.ylabel("Age")

plt.show()

# Middle-age customers churn more
# Older customers are more stable
```

```python
#ploting the boxplot of balance column with respect to churn

plt.figure(figsize=(6,4))

sns.boxplot(x='churn', y='balance', data=df)
# ploting the tiltle and lables of the boxplot
plt.title("Churn vs Balance")
plt.xlabel("Churn (0 = No, 1 = Yes)")
plt.ylabel("Balance")

plt.show()

# Low balance customers leave more
# High balance customers stay longer
```

```python
#ploting the count plot 
counts = df.groupby(['churn', 'products_number']).size().unstack()
#count ploting
counts.plot(kind='bar')

# Title aur labels
plt.title("Churn vs  products_number")
plt.xlabel("products_number")  
plt.ylabel("Count")

plt.show()

# Customers with 1 product churn more
# More products = more loyalty
```

```python
#ploting the count plot 
counts = df.groupby(['churn', 'active_member']).size().unstack()
#count ploting
counts.plot(kind='bar')

# Title aur labels
plt.title("Churn vs active_member")
plt.xlabel("active_member")  
plt.ylabel("Count")

plt.show()

# Inactive customers churn more
# Active users mostly stay
```

```python
#ploting the count plot 
counts = df.groupby(['churn', 'country']).size().unstack()
#count ploting
counts.plot(kind='bar')

# Title aur labels
plt.title("Churn vs Country")
plt.xlabel("Country")  
plt.ylabel("Count")

plt.show()

# Some countries have higher churn
# Location affects customer behavior
```

```python
#ploting the count plot 
counts = df.groupby(['gender', 'churn']).size().unstack()
#count ploting
counts.plot(kind='bar')

# Title aur labels
plt.title("Churn vs gender")
plt.xlabel("gender")  
plt.ylabel("Count")

plt.show()

# Slight difference in churn between different gender
```

```python
num_df = df.select_dtypes(include=['int64', 'float64'])
#  Calculate the correlation 
corr = num_df.corr()

# Step 4: Better heatmap plot karo
plt.figure(figsize=(10,8))  # increse the size of the heatmap

sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()
```

```python
# Customers with low balance churn more
# Customers with fewer products leave more
# Inactive users are more likely to churn
# Middle-age group shows higher churn
# Some regions have higher churn
```
